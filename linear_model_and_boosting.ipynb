# Метрика

Перед решением любой задачи важно понимать, как будет оцениваться ваше решение. В данном случае мы используем стандартную для задачи классификации метрику ROC-AUC. Ее можно вычислить, используя только предсказанные вероятности и истинные классы без конкретного порога классификации + она раотает даже если классы в данных сильно несбалансированны (примеров одного класса в десятки раз больше примеров длугого). Именно поэтому она очень удобна для соревнований.

Посчитать ее легко:


```python
from sklearn.metrics import roc_auc_score

y_true = [
    0,
    1,
    1,
    0,
    1
]

y_predictions = [
    0.1,
    0.9,
    0.4,
    0.6,
    0.61
]

roc_auc_score(y_true, y_predictions)
```

```python
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, RobustScaler, LabelEncoder, OneHotEncoder
from sklearn.pipeline import make_pipeline
```

```python
from sklearn.metrics import make_scorer
```

```python
#  Отнормируем числовие признаки
scaler = RobustScaler()
robust_scaled = scaler.fit_transform(data[num_cols])
robust_scaled_data = pd.DataFrame(robust_scaled, columns=scaler.get_feature_names_out(num_cols), index=data.index)
```

```python
# Категориальные признаки закодируем с помощью one-hot-encoding'а
encoder = OneHotEncoder(sparse_output=False)
one_hot_encoded = encoder.fit_transform(data[cat_cols])
one_hot_data = pd.DataFrame(one_hot_encoded, columns=encoder.get_feature_names_out(cat_cols), index=data.index)
```

```python
# Заменим данные после  one-hot-encoding'а
data_preparing = pd.concat([data, one_hot_data], axis=1)
data_preparing = data_preparing.drop(cat_cols, axis=1)
```

```python
# Заменим данные после нормирования
data_preparing = pd.concat([data_preparing, robust_scaled_data], axis=1)
data_preparing = data_preparing.drop(num_cols, axis=1)
```

```python
# Разбиваем данные на фичи и таргет
X = data_preparing.drop(columns=['Churn'])
y = data_preparing['Churn'].astype(int)
```

```python
# Делим выборку на train/valid
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.2, shuffle=True, random_state=21)
```

```python
# Обучаем модель
logistic_regression = LogisticRegressionCV(random_state = 42, solver = 'liblinear', penalty='l2', class_weight = 'balanced', scoring=make_scorer(roc_auc_score), refit=True)
logistic_regression.fit(X_train, y_train)
y_log_reg = logistic_regression.predict_proba(X_valid)[:,1]
```

```python
# Считаем ROC-AUC
roc_auc = roc_auc_score(y_valid, y_log_reg)
print("ROC-AUC Logistic Regression : %.4f" % roc_auc)
```

```python
from catboost import CatBoostClassifier
```

```python
# Разбиваем данные на фичи и таргет
X = data.drop(columns=['Churn'])
y = data['Churn'].astype(int)
```

```python
# Обучаем модель
catboost = CatBoostClassifier(random_state = 42, iterations=10, learning_rate=0.3, depth=6)
catboost.fit(X_train, y_train, cat_features=cat_cols)
y_catboost = catboost.predict_proba(X_valid)[:,1]
```

```python
# Считаем ROC-AUC
roc_auc = roc_auc_score(y_valid, y_catboost)
print("ROC-AUC Logistic Regression : %.4f" % roc_auc)
```
